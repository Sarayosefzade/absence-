```python
# convert it to a date-time
df["Date"] = pd.to_datetime(df["Date"], format = '%d/%m/%Y')
# display the first date
df["Date"][0]
```

```python
from datetime import datetime

year = df['Date'].dt.year
df.loc[:, "Year"] = year
month = df['Date'].dt.month
df.loc[:, "Month"] = month
day = df['Date'].dt.strftime('%A')
df.loc[:, "Day"] = day
df.drop("Date", axis=1)
df.head()
```

```python
education_encoding = {1: "high_school", 2: "graduate", 3: "postgraduate", 4: "master_phd"}
yes_no_encoding = {0: "No", 1: "Yes"}
df["Education_Level"] = df["Education_Level"].map(education_encoding)
df["Disciplinary_Incident"] = df["Disciplinary_Incident"].map(yes_no_encoding)
df["Social_Drinker_Status"] = df["Social_Drinker_Status"].map(yes_no_encoding)
df["Social_Smoker_Status"] = df["Social_Smoker_Status"].map(yes_no_encoding)
df.head()
```

```python
df["Absence_Reason"].value_counts().plot.bar()
```

سه دلیل اصلی غیبت به ترتیب مشاوره پزشکی، فیزیوتراپی و مشاوره دندان بوده

```python
def in_icd(val):
  if val > 21:
    return True
  else:
    return False
df["Disease"] = df["Absence_Reason"].apply(in_icd)
df["Disease"]
```

```python
df["Disease"].value_counts().plot.bar()
```

طبق داده‌های داده شده دلیل بیشتر از نیمی از غایبین مریضی بوده
 و در بازه آی سی دی قرار می‌گیرند

```python
smokers_count = df[df["Social_Smoker_Status"] == "Yes"].shape[0]
alcoholics_count = df[df["Social_Drinker_Status"] == "Yes"].shape[0]
total_records = df.shape[0]
smokers_percentage = (smokers_count / total_records) * 100
alcoholics_percentage = (alcoholics_count / total_records) * 100
print("Percentage of absence for smokers: {:.2f}%".format(smokers_count))
print("Percentage of absence for alcoholics: {:.2f}%".format(smokers_count))
```

پنجاه درصد سیگاری و پنجاه درصد مصرف کننده الکل بودند

```python
drinkers = df[df["Social_Drinker_Status"] == "Yes"]
drinkers_absence_counts = drinkers['Absence_Reason'].value_counts()
plt.figure(figsize=(10, 6))
drinkers_absence_counts.plot(kind='bar')
plt.xlabel('absence reason')
plt.ylabel('drinkers count')
plt.show()
```

بیشترین دلیل غیبت مصرف کنندگان الکل مشاوره دندان و مشاوره پزشکی بوده
بیش از هشتاد نفر از حدود صد نفری که به دلیل مشاوره دندان غایب بودن از افراد مصرف کننده الکل بوده

```python
drinkers = df[df["Social_Drinker_Status"] == "Yes"]
nondrinkers = df[df["Social_Drinker_Status"] == "No"]
combined_data = pd.concat([drinkers, nondrinkers])
plt.figure(figsize=(10, 6))
sns.violinplot(x='Social_Drinker_Status', y='Absence_Duration_Hours', data=combined_data)
plt.xlabel('drinker')
plt.ylabel('absence hours')
plt.show()
```

در بین مصرف کنندگان الکل برخی از افرادبیش از صد و بیست ساعت غیبت داشتن اما در
بین مصرف کنندگان سیگار زمان غیبت ها بیشتر از هشتاد ساعت نبوده همچنین تعداد افرادی که بین بیست تا چهل ساعت غیبت داشتن در الکلی ها بیشتر بود و افراد سیگاری به طور میانگین نسبت به سایرین ساعات غیبت کمتری داشتن
از پلات ویالونی نیز زمانی استفاده میکنیم که می‌خواهیم توزیع داده‌های عددی را مشاهده کنیم و در بازه‌های متفاوت میتوانیم به راحتی گروه‌های مختلف رو مقایسه کنیم

```python
smokers = df[df["Social_Smoker_Status"] == "Yes"]
nonsmokers = df[df["Social_Smoker_Status"] == "No"]
combined_data = pd.concat([smokers, nonsmokers])
plt.figure(figsize=(10, 6))
sns.violinplot(x='Social_Smoker_Status', y='Absence_Duration_Hours', data=combined_data)
plt.xlabel('smoker')
plt.ylabel('absence hours')
plt.show()
```


طبق جوینت پلات می‌توان دید که افرادی که هزینه حمل و نقل آنهابین صد و پنجاه دلار تا دویست دلار بوده بیشتر غیبت را داشتن
زمانی که میخواهیم بین دو متغیر متفاوت مقایسه ای انجام دهیم همچنین به توزیع آنها به صورت جداگانه نیز اهمیت میدهیم از جوینت پلات استفاده میکنیم

```python
from scipy import stats
transformed_absence_hours, lambda_ = stats.yeojohnson(df['Absence_Duration_Hours']+1)
df["Absence_Duration_Hours"] = transformed_absence_hours
sns.jointplot(data=df, x='Distance_to_Work', y='Absence_Duration_Hours', kind='scatter')
plt.xlabel('distance to work')
plt.ylabel('absence hours')
plt.show()
```

فاکتور فاصله تا محل کار با ساعت غیبت تقریبا نسبت مستقیم داشته و اکثرا ساعت غیبتشان یک و چهار دهم ساعت بوده
اما هزینه حمل و نقل ابتدا صعودی و سپس نزولی بود و باز هم در هزینه بین صد و پنجاه دلار تا دویست دلار بیشترین ساعات غیبت را دارند

```python
sns.jointplot(data=df, x='Transportation_Expense', y='Absence_Duration_Hours', kind='scatter')
plt.xlabel('Transportation Expense')
plt.ylabel('absence hours')
plt.show()
```
